```python
from pathlib import Path
import os
from collections import Counter, defaultdict
import numpy as np
from xml.etree import ElementTree as etree

from bs4 import BeautifulSoup, SoupStrainer, Tag, NavigableString
from itertools import takewhile, chain

import pandas as pd
import numpy as np

from nltk.corpus import stopwords
import re

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "xslt/output"
ext = "*.xml"
docTranscripts = [str(p) for p in Path(path).glob(ext)]
```

```python
def get_text(from_tag, until_tag):
    """ Parameters
        ----------
        from_tag : str
            Name of start element, e.g. "handShift".
        until_tag : str
            Name of next element until which to parse to, e.g. "handShift".
            
        Returns
        ----------
        word_dic : dict
            Dictionary with 'from_tag' as key and parsed text as value.
    """
    focus = SoupStrainer("sourceDoc")
    word_dic = defaultdict(list)
    
    for doc in docTranscripts:
        with open(doc) as fp:
            soup = BeautifulSoup(fp, "xml",parse_only=focus)

            for big in soup(from_tag):
                until = big.findNext(until_tag)
                strings = (node for node in big.nextSiblingGenerator() if getattr(node, 'text', '').strip())
                selected = takewhile(lambda node: node != until, strings)
                try:
                    for node in chain([big, next(selected)], selected):
                        if node.has_attr('hand'):
                            # catches text by different hand than last handShift
                            word_dic[node['hand']].append(node.text)
                        else:
                            word_dic[big['new']].append(node.text)
                except StopIteration as e:
                    pass
    return word_dic
```

```python
word_dic = get_text('handShift', 'handShift')
```

```python
def clean_text(word_dic, hand="g"):
    """ Parameters
        ----------
        word_dic : dict
            Dictionary as created by :func:get_text().
        hand : string
            Specifies which item in 'word_dic' to use.
        
        Returns
        ----------
        text : list
            List of cleaned strings written by 'hand'.
    """
    # Splits text written by "hand" and  removes escape sequences
    for i in list(word_dic.items()):
        if i[0]==hand:
            text = i[1]
            text = " ".join(text).split(" ")
            text = [re.sub('\s+', '', i.lower()) for i in text if i]
    return text

g_bl = clean_text(word_dic, "#g_bl")
g_t = clean_text(word_dic, "#g_t")
```

```python
def shorten_list(text, length=3):
    """ Parameter
        ---------
        text : list
            List of strings as created by :func:clean_text().
        length : int
            Minimum length of strings.
            
        Returns
        ---------
        short_text : list
            List of strings that are longer than or equally as long as 'length'.
    """
    short_text = []
    for i in text:
        if len(i)>=length:
            short_text.append(i)
    return short_text

g_bl = shorten_list(g_bl)
g_t = shorten_list(g_t)
```

```python
def remove_stop(text):
    # removes stopwords from list of strings
    stops = set(stopwords.words('german'))

    ohnestop = [word for word in text if word not in stops]
    return ohnestop

g_bl_ohne = remove_stop(g_bl)
g_t_ohne = remove_stop(g_t)
```

```python
def generate_count_df(word_list, hand):
    # creates pd.DataFrame from list of strings and their count    
    c = Counter(word_list)
    df = pd.DataFrame(c, index=[0]).transpose()
    df["Wörter"] = df.index
    df["Hand"]=hand
    df = df.rename(columns={0: "Häufigkeit"})
    df = df.sort_values(by="Häufigkeit", ascending=False)
    return df

g_bl_df = generate_count_df(g_bl, hand="g_bl")
g_bl_ohne_df = generate_count_df(g_bl_ohne, hand="g_bl_ohne")
g_t_df = generate_count_df(g_t, hand="g_t")
g_t_ohne_df = generate_count_df(g_t_ohne, hand="g_t_ohne")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Wörter", y="Häufigkeit", data=g_t_ohne_df[:10], color="g")
plt.savefig("tex/images/g_t_ohne_df.png")
```

```python
mit = pd.concat([g_bl_df,g_t_df]).sort_values(by='Häufigkeit', ascending=False)
ohne = pd.concat([g_bl_ohne_df,g_t_ohne_df]).sort_values(by='Häufigkeit', ascending=False)
```
